```python
def sum_numbers(num1, num2):
    if type(num1) == int:
        if type(num2) == int:
            return num1 + num2
    elif type(num1) == str:
        if type(num2) == str:
            return int(num1) + int(num2)
```

```python
sum_numbers('10', '20')
```
