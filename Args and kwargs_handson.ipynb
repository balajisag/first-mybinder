```python
def myfunc(a,b):
    # Returns 5% of the sum of a and b
    return sum((a,b)) * 0.05
```

```python
myfunc(40, 60)
```

```python
myfunc(40, 60, 80)
```

```python
def myfunc(a,b,c=0,d=0,e=0):
    # Returns 5% of the sum of a and b
    return sum((a,b,c,d,e,)) * 0.05
```

```python
myfunc(40, 60)
```

```python
myfunc(40, 60, 100, 100, 3)
```

```python
def myfunc(*args):
    return sum(args) * 0.05
```

```python
myfunc(40, 60, 100, 100, 3)
```

```python
myfunc(40, 60)
```

```python
def myfunc(*args):
    print(args)
```

```python
myfunc(40, 60, 100, 100, 3)
```

```python
def myfunc(*temp): # *args are nothing but argument but can be replaced by other names. But it is recommended to use args for others to be able to understand easily
    print(temp)
```

```python
myfunc(40, 60, 100, 100, 3)
```

```python
def myfunc(*args): 
    for x in args:
        print(x)
```

```python
myfunc(40, 60, 100, 100, 3)
```

```python
def myfunc(**kwargs):
    if 'fruit' in kwargs:
        print('My fruit of choice is {}'.format(kwargs['fruit']))
    else:
        print('I did not find any fruit here')
```

```python
myfunc(fruit='apple')
```

```python
myfunc(fruit='apple', veggie= 'lettuce')
```

```python
def myfunc(**kwargs):
    print(kwargs)
    if 'fruit' in kwargs:
        print('My fruit of choice is {}'.format(kwargs['fruit']))
    else:
        print('I did not find any fruit here')
```

```python
myfunc(fruit='apple', veggie= 'lettuce')
```

```python
def myfunc(*args, **kwargs):
    print('I would like {} {}'.format(args[0], kwargs['food']))
```

```python
myfunc(10,20,30,fruit='orange', food='eggs', animal='dog')
```

```python
def myfunc(*args, **kwargs):
    print(args)
    print(kwargs)
    print('I would like {} {}'.format(args[0], kwargs['food']))
```

```python
myfunc(10,20,30,fruit='orange', food='eggs', animal='dog')
```

```python
myfunc(10,20,30,fruit='orange', food='eggs', animal='dog', 100)
```

```python
myfunc(10,20,30, 100,fruit='orange', food='eggs', animal='dog')
```
