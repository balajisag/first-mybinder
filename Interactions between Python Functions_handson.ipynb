```python
def player_guess():
    guess=''
    while guess not in ['0', '1', '2']:
        guess = input("Pick a number: 0, 1, or 2:")
    return int(guess)
```

```python
def check_guess(mylist, guess):
    if mylist[guess] == 'X':
        print("Correct!")
    else:
        print("Wrong guess!")
        print(mylist)
```

```python
# Initial List
mylist = [' ', 'X', ' ']

# Shuffle List
mixedup_list = shuffle_list(mylist)

# User Guess
guess = player_guess()

# Check Guess
check_guess(mixedup_list, guess)
```
